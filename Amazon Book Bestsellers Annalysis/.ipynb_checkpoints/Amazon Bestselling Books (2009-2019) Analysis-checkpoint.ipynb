#  Load the Dataset and Perform Initial Exploration
### We'll use Pandas to load the dataset and examine its structure.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")


# Load the dataset
file_path = "bestsellers with categories.csv"  

df = pd.read_csv(file_path)

# Display first few rows
df.head()
```

# Check for Missing or Duplicate Data
### Before analysis, we need to ensure the dataset is clean.

```python
# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())

# Check for duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())

# Display dataset information
df.info()
```

### No missing values or duplicates found, but just incase let convert the Year collumn to categorical since we are not going to perform any mathematical operations on it and also to make it easier to handle with libraries like seaborn and  pandas.

# Data Cleaning
### We'll clean column names and convert YEAR coumn to categorical.

```python
# Rename columns for better readability 
df.columns = df.columns.str.strip().str.upper().str.replace(" ", "_")

# Convert year to categorical
df["YEAR"] = df["YEAR"].astype(str)

# Display updated dataset info
df.info()
```

# Summary Statistics & Data Distribution
### We'll analyze the key statistics of numerical columns.

```python
# Summary statistics for numerical columns
df.describe()


```

```python
# Count of books per year
plt.figure(figsize=(12, 6))
ax = sns.countplot(
    data=df, 
    x="YEAR", 
    palette="viridis", 
    order=sorted(df["YEAR"].unique())
)

# Add labels on top of bars
for p in ax.patches:
    ax.annotate(
        format(p.get_height(), '.0f'),  # Convert count to integer
        (p.get_x() + p.get_width() / 2., p.get_height()),  # Positioning
        ha='center', va='bottom', fontsize=11, fontweight='bold', color='black'
    )

# Improve readability
plt.title("Number of Bestselling Books per Year", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Count of Bestsellers", fontsize=12)
plt.xticks(rotation=45, fontsize=11)
plt.yticks(fontsize=11)

# Show grid for better visibility
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python
# Distribution of User Ratings
plt.figure(figsize=(10, 5))
sns.histplot(df['USER_RATING'], bins=10, kde=True, color="blue")
plt.title("Distribution of User Ratings", fontsize=14, fontweight='bold')
plt.xlabel("User Rating", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Distribution of Reviews
plt.figure(figsize=(10, 5))
sns.histplot(df['REVIEWS'], bins=20, kde=True, color="green")
plt.title("Distribution of Reviews", fontsize=14, fontweight='bold')
plt.xlabel("Number of Reviews", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Genre comparison
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="GENRE", palette="coolwarm")
plt.title("Distribution of Books by Genre", fontsize=14, fontweight='bold')
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

```python
# Top-performing authors
top_authors = df['AUTHOR'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(y=top_authors.index, x=top_authors.values, palette="magma")
plt.title("Top 10 Bestselling Authors", fontsize=14, fontweight='bold')
plt.xlabel("Number of Bestsellers", fontsize=12)
plt.ylabel("Author", fontsize=12)
plt.show()
```

```python
# Select only numerical columns
numeric_df = df.select_dtypes(include=["number"])

# Set figure size and styling
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Generate the heatmap with better aesthetics
heatmap = sns.heatmap(
    numeric_df.corr(), 
    annot=True, 
    cmap="coolwarm", 
    linewidths=1, 
    fmt=".2f",  # Limit decimal places to 2
    square=True, 
    cbar_kws={"shrink": 0.8}  # Adjust color bar size
)

# Improve title formatting
plt.title("Correlation Between Numerical Features", fontsize=16, fontweight="bold", pad=15)

# Show the heatmap
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Scatter plot 
sns.scatterplot(data=df, x="PRICE", y="USER_RATING", alpha=0.6, color="blue", s=80, edgecolor="black")

# Regression line for trend analysis
sns.regplot(data=df, x="PRICE", y="USER_RATING", scatter=False, color="red", line_kws={"linewidth": 2, "linestyle": "--"})

plt.title("Price vs. User Rating", fontsize=14, fontweight="bold")
plt.xlabel("Price ($)", fontsize=12)
plt.ylabel("User Rating", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Boxplot 
sns.boxplot(data=df, x="GENRE", y="USER_RATING", palette="coolwarm", width=0.6, notch=True, showmeans=True, 
            meanprops={"marker":"o", "markerfacecolor":"black", "markeredgecolor":"black", "markersize":"8"})

# Swarm plot for individual data points (avoid overlap using jitter)
sns.stripplot(data=df, x="GENRE", y="USER_RATING", color="black", alpha=0.5, jitter=True)

plt.title("User Rating Comparison: Fiction vs Nonfiction", fontsize=14, fontweight="bold")
plt.xlabel("Genre", fontsize=12)
plt.ylabel("User Rating", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
